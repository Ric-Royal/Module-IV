```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate Demand and Supply Curves
price = np.linspace(1, 10, 100)
demand = 110 - 10 * price
supply = -10 + 10 * price

# Find Equilibrium
equilibrium_price = np.where(np.isclose(demand, supply))[0][0]
equilibrium_quantity = demand[equilibrium_price]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(price, demand, label='Demand Curve', color='blue')
plt.plot(price, supply, label='Supply Curve', color='green')
plt.axvline(x=price[equilibrium_price], color='red', linestyle='--', label=f'Equilibrium Price = {price[equilibrium_price]:.2f}')
plt.axhline(y=equilibrium_quantity, color='orange', linestyle='--', label=f'Equilibrium Quantity = {equilibrium_quantity:.2f}')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.title('Supply and Demand Curves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate Demand and Supply Curves
price = np.linspace(1, 10, 100)
demand = 110 - 10 * price**2
supply = -10 + 10 * price**2

# Find Equilibrium
difference = np.abs(demand - supply)
equilibrium_price_index = np.argmin(difference)
equilibrium_price = price[equilibrium_price_index]
equilibrium_quantity = demand[equilibrium_price_index]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(price, demand, label='Demand Curve', color='blue')
plt.plot(price, supply, label='Supply Curve', color='green')
plt.axvline(x=equilibrium_price, color='red', linestyle='--', label=f'Equilibrium Price = {equilibrium_price:.2f}')
plt.axhline(y=equilibrium_quantity, color='orange', linestyle='--', label=f'Equilibrium Quantity = {equilibrium_quantity:.2f}')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.title('Supply and Demand Curves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the production function
def production_function(labor_input):
    return 100*labor_input - 2.5*labor_input**2 + 0.0025*labor_input**3

# Generate labor input values
labor = np.linspace(0, 40, 400)
output = production_function(labor)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(labor, output, label='Production Function', color='blue')

# Highlighting the stages
plt.axvspan(0, 20, color='green', alpha=0.1, label='Stage I: Increasing Returns')
plt.axvspan(20, 30, color='yellow', alpha=0.1, label='Stage II: Diminishing Returns')
plt.axvspan(30, 40, color='red', alpha=0.1, label='Stage III: Negative Returns')

plt.xlabel('Labor Input')
plt.ylabel('Total Output')
plt.title('Three Stages of Production')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define utility functions for goods A and B
def utility_A(quantity_A):
    return 50 * np.sqrt(quantity_A)

def utility_B(quantity_B):
    return 40 * np.sqrt(quantity_B)

# Prices for goods A and B
price_A = 5
price_B = 4

# Calculate marginal utility per dollar for a range of quantities
quantity_range = np.linspace(1, 20, 100)
mu_per_dollar_A = np.gradient(utility_A(quantity_range), quantity_range) / price_A
mu_per_dollar_B = np.gradient(utility_B(quantity_range), quantity_range) / price_B

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(quantity_range, mu_per_dollar_A, label='Marginal Utility per Dollar for Good A', color='blue')
plt.plot(quantity_range, mu_per_dollar_B, label='Marginal Utility per Dollar for Good B', color='green')

# Highlighting the equilibrium point
equilibrium_point = np.where(np.isclose(mu_per_dollar_A, mu_per_dollar_B))[0][0]
plt.axvline(x=quantity_range[equilibrium_point], color='red', linestyle='--', label=f'Equilibrium at Quantity = {quantity_range[equilibrium_point]:.2f}')

plt.xlabel('Quantity')
plt.ylabel('Marginal Utility per Dollar')
plt.title('Consumer Equilibrium')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate Data for Total Utility and Marginal Utility
quantities = np.linspace(0, 20, 100)
# A simple utility function: square root (representing diminishing MU)
total_utility = np.sqrt(quantities) * 10
marginal_utility = np.gradient(total_utility)

# 2. Visualize Total Utility and Marginal Utility
plt.figure(figsize=(12, 6))
plt.plot(quantities, total_utility, label='Total Utility', color='blue')
plt.plot(quantities, marginal_utility, label='Marginal Utility', color='green', linestyle='--')
plt.title('Total Utility and Marginal Utility of Apples')
plt.xlabel('Quantity of Apples')
plt.ylabel('Utility')
plt.legend()
plt.grid(True)
plt.show()

# 3. Generate Data for Indifference Curves
# Assuming a simple Cobb-Douglas utility function: U(A,B) = A^0.5 * B^0.5
A, B = np.meshgrid(np.linspace(0.1, 10, 100), np.linspace(0.1, 10, 100))
utility_levels = A**0.5 * B**0.5

# 4. Visualize Indifference Curves
plt.figure(figsize=(12, 6))
contour = plt.contour(A, B, utility_levels, levels=10, colors='blue')
plt.clabel(contour, inline=1, fontsize=10)
plt.title('Indifference Curves')
plt.xlabel('Quantity of Apples')
plt.ylabel('Quantity of Bananas')
plt.grid(True)
plt.show()
```

```python
import numpy as np

def bond_yield(price, par, coupon_rate, years_to_maturity):
    """
    Calculate the yield to maturity (YTM) for a bond.
    
    Parameters:
    - price: Current market price of the bond.
    - par: Par value of the bond.
    - coupon_rate: Annual coupon rate (as a fraction).
    - years_to_maturity: Years until the bond matures.
    
    Returns:
    - ytm: Yield to maturity.
    """
    # Initial guess for YTM
    ytm = coupon_rate
    previous_ytm = 0
    tolerance = 0.0001
    
    # Iteratively estimate YTM using the secant method
    while abs(ytm - previous_ytm) > tolerance:
        previous_ytm = ytm
        coupon = coupon_rate * par
        ytm = previous_ytm - (price - sum([coupon / (1 + previous_ytm)**i for i in range(1, years_to_maturity + 1)]) - par / (1 + previous_ytm)**years_to_maturity) / \
               (-sum([i * coupon / (1 + previous_ytm)**(i+1) for i in range(1, years_to_maturity + 1)]) + years_to_maturity * par / (1 + previous_ytm)**(years_to_maturity+1))
    
    return ytm

# Example
price = 950
par = 1000
coupon_rate = 0.05
years_to_maturity = 5
ytm = bond_yield(price, par, coupon_rate, years_to_maturity)
print(f"Yield to Maturity: {ytm*100:.2f}%")
```
