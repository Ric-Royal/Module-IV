## Question 2

2a) The data structure is a dictionary. It has 2 keys. These keys are Student A and Student B. Each key-value pair maps the key to its associated value.


```python
students = {'Student A': {'Math': 67, 'English': 78},'Student B': {'Math': 70, 'English': 70}}
print(students)

# list the keys in the dictionary
print(students.keys())
# list the values in the dictionary
print(students.values())
```

```python
# access the English grade for student B
print(students['Student B']['English'])
```

```python
# Change the English grade for student A to 90
students['Student A']['English'] = 90
print(students)
```

```python
# Add 'Student C' with Math grade of 80 and English grade of 99
students['Student C'] = {'Math': 80, 'English': 99}
print(students)
```

```python
# Function to check if a key exists in a dictionary
def check_key(key):
    if key in students.keys():
        print("Key exists")
    else:
        print("Key does not exist")
```

The big O complexity if the key does not exist is O(n)

```python
# from the students dictionary, write a function that returns the total for math and english for each student 
# and also return the average of the students for both math and english

def total_average():
    for key, value in students.items():
        print(key)
        print(value)
        print("Total: ", value['Math'] + value['English'])
        print("Average: ", (value['Math'] + value['English'])/2)

total_average()
```
